# Task 1: Exploring and Visualizing the Iris Dataset

## Objective
The objective of this task is to explore the Iris dataset using Exploratory Data Analysis (EDA). We will inspect the dataset, check for missing values, generate descriptive statistics, and create visualizations to understand the relationships between different flower measurements.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully!")
```

```python
# Load the Iris dataset directly from seaborn
df = sns.load_dataset('iris')

# Show the first 5 rows
print('First 5 rows:')
df.head()
```

```python
# How many rows and columns?
print('Shape:', df.shape)  # Output: (150, 5)

# What are the column names?
print('Columns:', df.columns.tolist())
```

```python
# What data types does each column have?
df.info()
```

```python
#Get statistical summary (min, max, average, etc.)
df.describe()
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

### Missing Value Analysis

The Iris dataset contains no missing values. Therefore, no data cleaning or imputation was required before visualization.,

```python
# How many flowers of each species?
df['species'].value_counts()
```

## Dataset Observation

- The dataset contains 150 rows and 5 columns.
- There are 4 numerical features and 1 categorical feature (species).
- No missing values are present.
- Each species contains 50 samples.

```python
# Scatter plot: Sepal Length vs Sepal Width
plt.figure(figsize=(8, 5))   # set chart size
sns.scatterplot(
    data=df,
    x='sepal_length',
    y='sepal_width',
    hue='species',   # color by species
    palette='Set1'
)
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

## Scatter Plot Analysis

- Setosa flowers form a clearly separate cluster.
- Versicolor and Virginica overlap slightly.
- Sepal measurements help distinguish Setosa from the other species.

```python
# Histogram for all 4 measurements
df.drop('species', axis=1).hist(
    bins=20,
    figsize=(10, 6),
    color='steelblue',
    edgecolor='black'
)
plt.suptitle('Distribution of All Features', y=1.02)
plt.tight_layout()
plt.show()
```

## Histogram Analysis

- Petal Length and Petal Width have distinct distributions.
- Sepal features have more overlap.
- Most measurements are approximately normally distributed.

```python
# Box plots for each feature grouped by species
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features):
    ax = axes[i // 2, i % 2]
    sns.boxplot(data=df, x='species', y=feature, hue='species', palette='pastel', legend=False, ax=ax)
    ax.set_title(f'Box Plot of {feature}')

plt.tight_layout()
plt.show()
```

## Box Plot Analysis

- Setosa has the smallest petal measurements.
- Virginica generally has the largest flowers.
- Only a few minor outliers are present.
- Petal measurements separate species better than sepal measurements.

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df.drop("species", axis=1).corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

## Heatmap Analysis

- Petal Length and Petal Width have the strongest positive correlation.
- Sepal Length has moderate correlation with petal measurements.
- Sepal Width has the weakest relationship with other features.

```python
sns.pairplot(
    df,
    hue="species",
    palette="Set2"
)

plt.show()
```

## Pairplot Analysis

- Pairplots compare every numerical feature against every other feature.
- Setosa is clearly separated from the other two species.
- Petal features provide the best separation for classification.
